```python
# Dependencies
import pandas as pd
import numpy as np
import json
import requests
from bs4 import BeautifulSoup
from splinter import Browser
#file path to saved csv dataframe
file = 'Cleaned_Data/main_city_shire_clean.csv'
bom_file = 'Raw_Data/bom_table.csv'
```

```python
#bring csv file in as pd dataframe
cities_df = pd.read_csv(file, encoding="ISO-8859-1")
cities_df.head()
```

```python
bom_df = pd.read_csv(bom_file, encoding="ISO-8859-1")
bom_df
```

```python
# make site refs a list
sites = bom_df["site_ref"].tolist()
refs = np.array(sites)
print(refs)
```

```python
print(len(sites))
```

```python
#loop through sites to create urls to scrape mean high temp 
temps = []

for site in refs:
    site = site
    url = f'http://www.bom.gov.au/climate/averages/tables/cw_0{site}.shtml'
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    high_temp = soup.find("td", class_='highest').text
    temps.append(high_temp)
    #time.sleep(3.01)
print(temps)    
```

```python
print(len(temps))
```

```python
# add mean high temp to bom df
bom_df["mean_high_temp"]=temps
bom_df
```

```python
# Rename column City in bom_df
bom_df = bom_df.rename(columns={"ï»¿City": "city", "bom_city":"bom_site"})
bom_df
```

```python
# merge cities_df with bom_df
city_temps_df = cities_df.merge(left_on= 'City', right= bom_df, right_on='city', how="inner")
city_temps_df
```

```python
#save city_temps df to csv
city_temps_df.to_csv("Cleaned_Data/city_temps.csv", encoding="utf-8", index=False)
```
